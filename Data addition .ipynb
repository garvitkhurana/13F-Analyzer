```python
import glob
import pandas as pd
```

```python
csv_list = glob.glob("/content/drive/My Drive/Analytics in Practice APG/Data/Excel files/*.csv")
csv_list
```

```python
dataset = pd.DataFrame()
for i in csv_list:
  Manager = str(i)[67:-4]
  df = pd.read_csv(i)
  df['CIK'] = Manager
  dataset = dataset.append(df,ignore_index=True)
```

```python
scrap_df = pd.read_excel("/content/drive/My Drive/Analytics in Practice APG/Data/manager_list.xlsx", converters={'CIK': str})
scrap_df['CIK'] = scrap_df['CIK'].astype(str)
scrap_df = scrap_df[['CIK', 'Manager Name']]
dataset['CIK'] = dataset['CIK'].astype(str)
new_df = pd.merge(dataset, scrap_df, on ='CIK', how ='inner')
new_df = new_df.set_index('Manager Name')
new_df.to_csv('/content/drive/My Drive/Analytics in Practice APG/Data/data_all.csv')
new_df
```
