```python
table_import = pd.read_excel("/content/drive/My Drive/Data_APG/data_with_ticker.xlsx")
table_import
```

```python
table_import['Date_D'] = pd.to_datetime(table_import['Date'])
table_import['Return_date'] = table_import['Date_D'] + timedelta(days=90)
table_import['Return_date'] = [x if x.weekday() not in [5,6] else x - timedelta(days = (x.weekday()-4)) for x in pd.to_datetime(table_import['Return_date'])]
table_import
```

```python
table_import['New_ticker'] = table_import['TICKER'].apply(lambda x: x.split(' ')[0] if x != 'No identifier found.' else x)
table_import['New_ticker'] = table_import['New_ticker'].apply(lambda x: x.split('/')[0] if x != 'No identifier found.' else x)
table_import['New_ticker'] = table_import['New_ticker'].apply(lambda x: x[:-3] if x.endswith('EUR') else x)
table_import['New_ticker'] = table_import['New_ticker'].apply(lambda x: x[:-3] if x.endswith('USD') else x)
table_import
```

```python
for ticker,date in zip(table_import['New_ticker'], table_import['Return_date']):
  try:
    stock_table_addition = yf.download('{}'.format(ticker), start=date, end=date + timedelta(days=1), progress=False)
    stock_table_addition['TICKER'] = ticker
    print(stock_table_addition['Adj Close'])
  except Exception as e: 
    print(e)
    pass
```

```python
table_import.sort_values('Adj Close')
table_import
```

```python
list_ticker = table_import['TICKER']
list_ticker_unique = list(set(list_ticker))
list_ticker_unique = list_ticker_unique
list_ticker_unique = [x for x in list_ticker_unique if not any(c.isdigit() for c in x)]
len(list_ticker_unique)
print(list_ticker_unique)
```

```python
new_list = []
for ticker in list_ticker_unique:
    if ticker.endswith('EUR'):
        ticker = ticker[:-3]
    elif ticker.endswith('USD'):
        ticker = ticker[:-3]
    elif ticker != 'No identifier found.':
        ticker = ticker.split(' ')
        ticker = ticker[0]
        ticker = ticker.split('/')
        ticker = ticker[0]
    new_list.append(ticker)
```

```python
import yfinance as yf
fundamentals_table_addtion = yf.download('{}'.format('WSC'), start='2021-10-19', end='2021-10-20', progress=False)
fundamentals_table_addtion
```

```python
stock_table= pd.DataFrame()

for ticker in new_list:
  try:
    stock_table_addition = yf.download('{}'.format(ticker), start='2021-10-19', end='2021-10-19', progress=False)
    stock_table_addition['TICKER'] = ticker
    stock_table = stock_table.append(stock_table_addition)
    print(stock_table)
  except Exception as e: 
    print(e)
    pass
```
