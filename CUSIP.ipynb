```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = '/content/drive/MyDrive/Data/data.csv'
```

```python
df=pd.read_csv(data_path)
df=df.sort_values('NAME OF ISSUER')
df.head()
```

```python
df=df[['Manager','NAME OF ISSUER','VALUE (x$1000)']]
```

```python
fig, ax1 = plt.subplots(figsize=(20, 20))
sns.barplot(x='VALUE (x$1000)',y='Manager',data=df, ax=ax1)
sns.despine(fig)
```

```python
fig, ax1 = plt.subplots(figsize=(20, 20))
sns.barplot(x='NAME OF ISSUER',y='VALUE (x$1000)',data=df, ax=ax1)
sns.despine(fig)
```

```python
df_issuer=pd.DataFrame(set(df['NAME OF ISSUER']),columns=['NAME OF ISSUER'])
df_issuer=df_issuer.sort_values('NAME OF ISSUER')
df_issuer
```

```python
df_cusip=pd.DataFrame(pd.DataFrame(df_issuer['NAME OF ISSUER']).merge(df[['CUSIP','NAME OF ISSUER']])).drop_duplicates()
df_cusip
```

```python
df_cusip.to_csv('/content/drive/MyDrive/Data/cusip.csv')
```
